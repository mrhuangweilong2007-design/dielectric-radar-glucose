```python
import numpy as np
import matplotlib.pyplot as plt

def debye_model(f, eps_s, eps_inf, tau):
    omega = 2 * np.pi * f
    return eps_inf + (eps_s - eps_inf) / (1 + 1j * omega * tau)

EPS_S_AIR = 78.4
EPS_INF_AIR = 5.2
TAU_AIR = 8.27e-12

frekuensi = np.logspace(8, 11, 500)

glucose_levels = [0, 50, 100, 150, 200]

plt.figure(figsize=(12, 6))

for glucose in glucose_levels:
    eps_s = EPS_S_AIR - (glucose / 100) * 1.5
    hasil = debye_model(frekuensi, eps_s, EPS_INF_AIR, TAU_AIR)

    plt.plot(
        frekuensi / 1e9,
        hasil.imag,
        label=f'{glucose} mg/dL',
        linewidth=2
    )

plt.xlabel('Frekuensi Radar (GHz)', fontsize=13)
plt.ylabel('ε" (Imaginary Part / Energy Loss)', fontsize=13)
plt.title('Efek Kadar Glukosa terhadap Penyerapan Gelombang Mikro (Model Debye)', fontsize=14)
plt.xscale('log')
plt.grid(True, which='both', linestyle='--', alpha=0.5)
plt.legend(title='Kadar Glukosa Darah', fontsize=11, title_fontsize=12)
plt.tight_layout()

plt.savefig('debye_simulation_graph.png', dpi=300)
plt.show()

print("✅ Simulasi Selesai! File 'debye_simulation_graph.png' telah dibuat.")
```
